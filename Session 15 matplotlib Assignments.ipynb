```python
fig, ax = plt.subplots(figsize=(10, 6))
years = df.columns[1:]
x = np.arange(len(df['batsman']))
width = 0.2

# Plot bars for each year
for i, year in enumerate(years):
    ax.bar(x + i * width, df[year], width, label=year)

ax.set_xlabel("Batsman")
ax.set_ylabel("Runs")
ax.set_title("Batsmen Performance Over the Years")
ax.set_xticks(x + width)
ax.set_xticklabels(df['batsman'], rotation=30, ha="right")
ax.legend(title="Year")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
years = df.columns[1:] # Years 2015, 2016, 2017
x = np.arange(len(df['batsman']))  # Batsman positions

# Plot stacked bars
bottom = np.zeros(len(df['batsman']))
for year in years:
    ax.bar(df['batsman'], df[year], label=year, bottom=bottom)
    bottom += df[year]

ax.set_xlabel("Batsman")
ax.set_ylabel("Total Runs")
ax.set_title("Cumulative Performance of Batsmen Over the Years")
ax.legend(title="Year")

plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

Task:
Create a pie chart representing the proportion of total runs contributed by the top 5 batsmen in the dataset. (batter.csv)

```python
top_batsmen = df.nlargest(5, 'runs')

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

plt.figure(figsize=(8, 8))
plt.pie(
    top_batsmen['runs'], 
    labels=top_batsmen['batter'], 
    autopct='%1.1f%%', 
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)

plt.title("Top 5 Batsmen Contribution to Total Runs", fontsize=14)

plt.show()
```

```python
top_batsmen = df.nlargest(5, 'strike_rate')

max_index = top_batsmen['strike_rate'].idxmax()

explode = [0.1 if i == max_index else 0 for i in top_batsmen.index]

plt.figure(figsize=(8, 8))
plt.pie(
    top_batsmen['strike_rate'], 
    labels=top_batsmen['batter'], 
    autopct='%1.1f%%', 
    colors=colors, 
    explode=explode,
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)

plt.title("Top 5 Batsmen by Strike Rate", fontsize=14)

plt.show()
```

```python
top_batsmen = df.nlargest(5, 'avg')

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

plt.figure(figsize=(8, 8))
plt.pie(
    top_batsmen['avg'], 
    labels=top_batsmen['batter'], 
    autopct=lambda p: '{:.2f}'.format((p/100)*top_batsmen['avg'].sum()),
    colors=colors, 
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)

plt.title("Top 5 Batsmen by Batting Average", fontsize=14)

plt.show()
```

```python
top_batsmen = df.nlargest(5, 'runs')

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Pie Chart 1: Total Runs
axes[0].pie(
    top_batsmen['runs'], 
    labels=top_batsmen['batter'], 
    autopct='%1.1f%%', 
    colors=colors, 
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)
axes[0].set_title("Top 5 Batsmen by Runs")

# Pie Chart 2: Strike Rate
axes[1].pie(
    top_batsmen['strike_rate'], 
    labels=top_batsmen['batter'], 
    autopct='%1.1f%%', 
    colors=colors, 
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)
axes[1].set_title("Top 5 Batsmen by Strike Rate")

# Pie Chart 3: Batting Average
axes[2].pie(
    top_batsmen['avg'], 
    labels=top_batsmen['batter'], 
    autopct='%1.1f%%', 
    colors=colors, 
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)
axes[2].set_title("Top 5 Batsmen by Average")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Scatter plot (total_bill vs tip)
sns.scatterplot(x="total_bill", y="tip", data=tips, ax=axes[0, 0])
axes[0, 0].set_title("Total Bill vs Tip")

# Bar plot (average tip per day)
sns.barplot(x="day", y="tip", data=tips, estimator=sum, ax=axes[0, 1])
axes[0, 1].set_title("Total Tip Amount Per Day")

# Histogram of size
sns.histplot(tips["size"], bins=6, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Histogram of Party Size")

# Box plot of total_bill
sns.boxplot(y="total_bill", data=tips, ax=axes[1, 1])
axes[1, 1].set_title("Box Plot of Total Bill")

plt.tight_layout()
plt.show()
```

```python
# Convert 'day' to categorical type for proper ordering
tips["day"] = pd.Categorical(tips["day"], categories=["Thur", "Fri", "Sat", "Sun"], ordered=True)

# Sort dataset by day for cumulative sum
tips_sorted = tips.sort_values(by="day")

# Calculate cumulative total_bill
tips_sorted["cumulative_total_bill"] = tips_sorted["total_bill"].cumsum()

# Create 2x1 subplots with shared x-axis
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Bar plot: Average total_bill per day
sns.barplot(x="day", y="total_bill", data=tips, estimator="mean", ax=axes[0])
axes[0].set_title("Average Total Bill per Day")
axes[0].set_ylabel("Average Total Bill")

# Line plot: Cumulative total_bill over time
sns.lineplot(x="day", y="cumulative_total_bill", data=tips_sorted, marker="o", ax=axes[1])
axes[1].set_title("Cumulative Total Bill Over Time")
axes[1].set_ylabel("Cumulative Total Bill")

# Improve layout
plt.xlabel("Day")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Scatter plot: Total Bill vs Tip
sns.scatterplot(x="total_bill", y="tip", data=tips, ax=axes[0], color="blue")
axes[0].set_title("Total Bill vs Tip")
axes[0].set_xlabel("Total Bill")
axes[0].set_ylabel("Tip (Shared Y-Axis)")

# Scatter plot: Total Bill vs Size
sns.scatterplot(x="total_bill", y="size", data=tips, ax=axes[1], color="green")
axes[1].set_title("Total Bill vs Size")
axes[1].set_xlabel("Total Bill")
axes[1].set_ylabel("Size (Shared Y-Axis)")

plt.tight_layout()
plt.show()
```
