```python
import matplotlib.pyplot as plt
import numpy as np
```

## Legends

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend()
```

```python
plt.plot(x, x**2, label = "x**2")
plt.plot(x, x**3, label = "x**3")
plt.legend()
plt.show()
```

```python
# ax.legend(loc = 1) # upper right corner
# ax.legend(loc = 2) # upper left corner
# ax.legend(loc = 3) # lower left corner
# ax.legend(loc = 4) # lower right corner

ax.legend(loc = 0) # let matplotlib decide the optimal location
fig
```

## Setting colors, linewidths, linetypes

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color='blue', alpha=0.5) # half-transparent
ax.plot(x, x+2, color='#8B008B') # RGB hex code
ax.plot(x, x+3, color='#FF8C00') # RGB hex code
```

### Line and Marker styles
### Line Width

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color='red', linewidth=0.25)
ax.plot(x, x-2, color='red', lw=0.50)
ax.plot(x, x-3, color='red', lw=1)
ax.plot(x, x-4, color='red', lw=10)
```

### Linestyles

```python
# possible linestyle options '--', '-', '-.', ':', 'steps'
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color='green', lw=3, linestyle='-') # solid
ax.plot(x, x-2, color='green', lw=3, ls='-.') # dash dot
ax.plot(x, x-3, color='green', lw=3, ls=':') # dots
ax.plot(x, x-4, color='green', lw=3, ls='--') # dashes
```

## Custom linestyle dash

```python
fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x,x)

print(type(lines))
```

```python
fig, ax = plt.subplots(figsize=(12,6))
#custom dash
lines = ax.plot(x, x+8, color='black', lw=5)
lines[0].set_dashes([10, 5]) # format: line length, space length
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# custom dash
lines = ax.plot(x, x+8, color='black', lw=5)
lines[0].set_dashes([1,1,1,1,10,10,5,5]) # format: line length, space length
```

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(12, 6))

# Plot multiple lines
lines = ax.plot(x, x+8, x, x+4, x, x)  # Three lines plotted

# Custom dash pattern for the first line
lines[0].set_dashes([1, 1, 1, 1, 10, 10, 5, 5])  

# Custom dash pattern for the second line
lines[1].set_dashes([5, 2, 10, 2])  

# Custom dash pattern for the third line
lines[2].set_dashes([2, 4, 8, 4])  

plt.show()
```

## Markers
## Markers Style

```python
fig, ax = plt.subplots(figsize=(12,6))

# use marker for string code
# use markersize or ms for size

ax.plot(x, x-1, marker='+', markersize=20)
ax.plot(x, x-2, marker='o', ms=20) # ms can be used for markersize
ax.plot(x, x-3, marker='s', ms=10, lw=1)
ax.plot(x, x-4, marker='1', ms=20)
```

## Custom marker edges, thickness, size and style

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x, color='black', lw=1, ls='-', marker='s', markersize=20,
       markerfacecolor='red', markeredgecolor='blue', markeredgewidth=8)
```

## Barplots

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

### Horizontal Bar plot

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

### Adding color

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()

#same methods like hex code, color names can be used here as well
```

### Bar width and height

```python
# The bar() takes the keyword argument width to set the width of the bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
# The barh() takes the keyword argument height to set the height of the bars:

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```
