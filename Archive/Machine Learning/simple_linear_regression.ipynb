# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset.head()
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Salary (y_test)")
plt.ylabel("Predicted Salary (y_pred)")
plt.title("Actual vs Predicted Salary")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
# Create a DataFrame to show y_test and y_pred side by side
comparison_df = pd.DataFrame({'Actual Salary (y_test)': y_test, 'Predicted Salary (y_pred)': y_pred})

print(comparison_df)
```

```python
def predict_salary(years_experience, regressor):

    predicted_salary = regressor.predict([[years_experience]])  # Input must be 2D array

    # Return the final predicted salary
    return predicted_salary[0]  # Return a scalar value

# Example usage:
years_of_experience = 5  # Change this value as needed
predicted_salary = predict_salary(years_of_experience, regressor)
print(f"Predicted salary for {years_of_experience} years of experience is: {predicted_salary:.2f}")
```
