## Boxplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

tips['total_bill']
```

```python
# Create a simple boxplot
sns.boxplot(x=tips["total_bill"])

plt.show()
```

### Boxplot for Categorical Data

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

### Adding hue for Grouping

```python
sns.boxplot(x="day", y="total_bill", hue='sex', data=tips)
plt.show()
```

### Adjusting Outliers and Whiskers (fliersize, showfliers)

```python
sns.boxplot(x="day", y="total_bill", data=tips, showfliers=False)  # Removes outliers
plt.show()
```

```python
# Outliers are hidden (showfliers=False), making the plot cleaner.
# You can control the outlier dot size using fliersize
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, fliersize=3)  # Smaller outliers
```

#### Changing Whisker Range (whis)

#### Control Over Data Spread Representation

If the default 1.5 × IQR hides important data trends, adjusting whis helps.

Example: Financial data may have extreme values that shouldn't be classified as outliers.

```python
sns.boxplot(x="day", y="total_bill", data=tips, whis=2)  # Whiskers extend to 2 × IQR
plt.show()
```

#### Customizing colors

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue='sex', palette="coolwarm")
plt.show()
```

### Horizontal Boxplots

```python
# To flip the boxplot horizontally, just swap x and y:
sns.boxplot(y="day", x="total_bill", data=tips)
plt.show()
```
