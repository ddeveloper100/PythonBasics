# plt.subplots()

#### NOTE: Make sure you put the commands all together in the same cell as we do in this notebook
The plt.subplots() object will act as a more automatic axis manager. This makes it much easier to show multiple plots side by side

Note how we use tuple unpacking to grab both the Figure object and a numpy array of axes

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

## Adding rows and columns

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows = 1 , ncols = 2)
```

## Plot on axes objects

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)

for axe in axes:
    axe.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
# Returns 2 dimensional array
fig, axes = plt.subplots(nrows =2, ncols =2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

```python
# We can use fig.tight_layout() ot plt.tight_layout() to automatically adjust the positions of the axes on the figure
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)

plt.tight_layout()
```

## Parameters on subplots()

Recall we have both the Figure object and the axes. Meaning we can edit properties at both levels.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')

axes[1][1].plot(x,y)
axes[1][1].set_title('1 1 Title')
axes[1][1].set_xlabel('1 1 X Label')

axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# Parameters at the figure level
fig.suptitle('Figure Level', fontsize=16)

plt.show()
```

## Manual Spacing on subplots()

```python
fig, axes = plt.subplots(nrows = 2 , ncols = 2 , figsize = (12,8))

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.subplots_adjust(left = None,
                   bottom = None, 
                   right = None, 
                   top = None, 
                   wspace = 0.5,
                   hspace = 0.3,)

plt.show()
```

## Exporting plt.subplots()

```python
# THis returns a 2 dimensional array and saves it

fig, axes = plt.subplots(nrows = 2 , ncols = 2 , figsize = (12,8))

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.savefig('subplots.png', bbox_inches = 'tight')

plt.show()
```
