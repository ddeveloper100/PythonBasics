#### 1. Handling Missing Values:
#### o Fill missing values in the 'reviews_per_month' column with the median of the column.

```python
df['reviews_per_month'].isnull().sum()
```

#### 3. Aggregating Data:
#### o Aggregate the data to find the minimum, maximum, and average prices for each 'room_type'.

```python
df['room_type'].unique()
```

```python
minimum_price = df.groupby('room_type')['listing_price'].min()
minimum_price
```

```python
maximum_price = df.groupby('room_type')['listing_price'].max()
maximum_price
```

```python
average_price = df.groupby('room_type')['listing_price'].mean()
average_price
```

```python
aggregated_data = pd.DataFrame({'Minimum Price': minimum_price, 'Maximum Price': maximum_price, 'Average Price': average_price})
aggregated_data
```

#### 9. Find Outliers:
#### o Identify listings with a price more than 3 standard deviations above the mean price.

```python
df_stddev = df['listing_price'].std()
df_mean = df['listing_price'].mean()
print(df_mean)
print(df_stddev)
print(df_mean+3*df_stddev)
 # > m+3*std
temp_res = df[df['listing_price']> df_mean+3*df_stddev]
temp_res['listing_price']
```

#### Merging Datasets

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Score': [85, 90, 75]
})
```

```python
df1
```

```python
df2
```

```python
merged_df = pd.merge(df1, df2, on='ID', how='inner')
```

```python
merged_df
```
