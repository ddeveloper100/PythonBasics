#### Stacked Bar Plot: Display the number of wins, draws, and losses for all home teams.

```python
df
```

```python
df.columns
```

```python
# detemining the win, loss or draw in the match
df['home_win'] = df['home_score'] > df['away_score']
df['home_loss'] = df['home_score'] < df['away_score']
df['home_draw'] = df['home_score'] == df['away_score']

results = df.groupby('home_team')[['home_win', 'home_draw', 'home_loss']].sum()

fig, ax = plt.subplots(figsize=(12,6))
results.plot(kind='bar', stacked=True, ax=ax, color=['green', 'yellow', 'blue'], edgecolor='black')

plt.title('Home Team Wins, Losses, and Draws')
plt.xlabel('Home Team Name')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.legend(['Wins', 'Losses', 'Draws'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

#### Pie Chart : Show the number of times country becomes runner up. (fifa_matches dataset)

```python
df['Round'].unique()
```

```python
temp_df = df[['home_team', 'away_team', 'Round', 'home_total_score', 'away_total_score']]
temp_df = temp_df[df['Round'] == 'Final']

temp_df.loc[temp_df['home_total_score']>temp_df['away_total_score'], 'runner_up_team'] = temp_df['away_team']
temp_df.loc[temp_df['home_total_score']<temp_df['away_total_score'], 'runner_up_team'] = temp_df['home_team']

temp_df = temp_df.groupby('runner_up_team').count()

plt.pie(temp_df['Round'], labels=temp_df.index, autopct='%0.1f%%')
plt.show()
```

#### Scatter Plot: Plot Birth_Weight against Birth_Length to visualize any relationship between the two variables. . (covid dataset) 

```python
covid_df['Birth_Length'] = covid_df['Birth_Length'].fillna(0)
covid_df['Birth_Weight'] = covid_df['Birth_Weight'].fillna(0)

sns.scatterplot(data=covid_df, x='Birth_Length', y='Birth_Weight')
plt.show()
```

#### 3D Scatter Plot: Create a 3D scatter plot to visualize the relationship between Maternal_Age, Gestational_Age_At_Birth, and Birth_Weight. . (covid dataset)

```python
covid_df['Maternal_Age'] = covid_df['Maternal_Age'].fillna(0)
covid_df['Gestational_Age_At_Birth'] = covid_df['Gestational_Age_At_Birth'].fillna(0)
covid_df['Birth_Weight'] = covid_df['Birth_Weight'].fillna(0)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(projection='3d')

ax.scatter(covid_df['Maternal_Age'], covid_df['Birth_Weight'], covid_df['Gestational_Age_At_Birth'])
ax.set_xlabel('Maternal_Age')
ax.set_ylabel('Gestational_Age_At_Birth')
ax.set_zlabel('Birth_Weight')
```

#### Subplots with Categorical Hue: Create subplots that compare Maternal_Age, Birth_Length, and Birth_Weight while coloring points based on Household_Income categories. (covid dataset)

```python
fig, axes = plt.subplots(3, 1, figsize=(15, 15))
sns.scatterplot(ax=axes[0], data=covid_df, x='Maternal_Age', y='Birth_Length', hue='Household_Income')
sns.scatterplot(ax=axes[1], data=covid_df, x='Birth_Length', y='Birth_Weight', hue='Household_Income')
sns.scatterplot(ax=axes[2], data=covid_df, x='Birth_Weight', y='Maternal_Age', hue='Household_Income')
plt.show()
```
